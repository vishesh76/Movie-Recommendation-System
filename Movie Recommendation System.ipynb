```python
# Importing libraries 
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
# Merging two data frame with respect to title
movies = movies.merge(credits,on='title')
```

```python
movies.head(5)
```

```python
# Choosing the variables that are important for our recommendation system and dropping the remaining variables
# genres,id,keywords,title,overview,cast,crew - we will be using these variables to build our model and will drop the remaining variables

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# Checking for missing values
movies.isnull().sum()
```

```python
# Dropping the missing values
movies.dropna(inplace=True)
```

```python
# Checking for duplicate values
movies.duplicated().sum()
```

```python
import ast
```

```python
# Preprocessing of Data
def convert (obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

This preprocessing of data is being done because the data saved in "genres" and "keywords" are in dictionary format. Hence to remove the id of those data and saving in the format string this operation is being done. 

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3 (obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

For the "cast" we are choosing only the first 3 actors or actress names.

```python
def fetch_director (obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

For the "crew" part, the name of director is only important factor for the recommendation system. Hence only choosing "Director"

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Transformation of dataset to remove the spaces between the strings
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Creating a new variable "tags" by concatenating other vairables.
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew'] 
```

```python
movies.head()
```

```python
# Choosing only 3 variables i.e., 'movie_id', 'title', 'tags'. As we have concatenated overview, genres, keywords, cast and crew into tags.
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# Converting tags(text) into vectors using Bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
# Calculating distance between vectors using Cosine distance as the data has high dimensionality measure
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
